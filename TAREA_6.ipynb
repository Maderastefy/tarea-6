```python
#PREGUNTA 2

#LIBRERIAS

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#definir la ecuacion diferencial
def equation (T, t, k):
   dTdt = k * (T-Tm)
   return dTdt

#parametros
Tm = 70
k = -0.19018
T0 = 300
t = np.linspace(0, 100, 100) #array de tiempo

#Resolver la ecuacion diferencial utilizando el metedo de Runge kutta
Solution = odeint(equation, T0, t, args=(k,))

#graficar
plt.plot(t, Solution, label='T')
plt.title('Grafica')
plt.xlabel('Tiempo')
plt.ylabel('Temperatura')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#definir la ecuacion
def equiation(T, t, k, Tm):
  dTdt = k * (T - Tm)
  return dTdt

#parametros
Tm = 70
T0 = 300
k = -0.19018
t = np.linspace(0, 100, 100)

#resolver la ecuacion diferencial usando odeint
Solution = odeint(equiation, T0, t, args=(k, Tm))

#graficar
plt.plot(t, Solution, label='T')
plt.title('Enfriamiento')
plt.xlabel('tiempo')
plt.ylabel('temperatura')
plt.legend()
plt.show()
```

```python
#Pregunta 3

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

#Definir las ecuaciones diferenciales del modelo
def model(t, y, k1, k2, k3, k4, k5):
  dxdt = k1 * y[0] - k2 * y[0]**2 - k3 * y[0] * y[1]
  dydt = k4 * y[1] - k2 * y[1]**2 - k5 * y[0] * y[1]
  return [dxdt, dydt]

#parametros
k1 = 1
k2 = 0.1
k3 = 0.05
k4 = 1.7
k5 = 0.15

#condiciones iniiales y tiempo de integracion
initial_condictions = [1.1, 1] #poblacion inicial p y d
time_span = (0, 200) #tiempo de integracion

#resolver las ecuacion diferenciales
solution = solve_ivp(model, time_span, initial_condictions, args=(k1, k2, k3, k4, k5), dense_output=True)

#Evaluar la solucion de un conjunto de tiempo mas fino
t_fine = np.linspace(0, 100, 1000)
y_fine = solution.sol(t_fine)

#graficar las poblaciones
plt.figure(figsize=(10, 6))
plt.plot(t_fine, y_fine[0], label='x(t)')
plt.plot(t_fine, y_fine[1], label='y(t)')
plt.title('Modelo')
plt.xlabel('tiempo')
plt.ylabel('Poblacion')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Pregunta 3. Otras condiciones

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

#Definir las ecuaciones
def MODEL(t, Y, m1, m2, m3, m4, m5):
  dXdt = m1 * Y[0] - m2 * Y[0]**2 - m3 * Y[0] * Y[1]
  dYdt = m4 * Y[1] - m2 * Y[1]**2 - m5 * Y[0] * Y[1]
  return (dXdt, dYdt)

#Parametros
m1 = 1
m2 = 0.1
m3 = 0.05
m4 = 1.7
m5 = 0.15

#condiciones iniciales
Cond_inicial = [4, 10]
Tiempo = (0, 200)

#resolver
Sol = solve_ivp(MODEL, Tiempo, Cond_inicial, args=(m1, m2, m3, m4, m5), dense_output=True)

#evaluar
t_fino = np.linspace(0, 100, 1000)
Y_fine = Sol.sol(t_fino)

#graficar
plt.figure(figsize=(10, 6))
plt.plot(t_fino, Y_fine[0], label='X(t)')
plt.plot(t_fino, Y_fine[1], label='Y(t)')
plt.title('MODELO 2')
plt.xlabel('Tiempo')
plt.ylabel('Poblacion')
plt.legend()
plt.grid(True)
plt.show()
```
